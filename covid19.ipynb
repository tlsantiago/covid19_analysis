```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('C:/Users/ssant/Desktop/Data/covid19_analysis/general.csv')
```

```python
# Changing the datatype and replacing commas with dots

df.dtypes

df['human_development_index'] = df["human_development_index"].str.replace(',','.')

df['human_development_index'] = df.human_development_index.astype(float)
```

```python
# Function to identify ONU's HDI tier

def HDITier(x):
    
    if x >= 0.8 and x <= 1.0:
        return "Very High Human Development"
    elif x >= 0.7 and x <= 0.79:
        return "High Human Development"
    elif x >= 0.55 and x <= 0.69:
        return "Medium Human Development"
    else:
        return "Low Human Development"
```

```python
for row in df.iterrows():
    df['HDI_tier'] = df.apply(lambda x: HDITier(x['human_development_index']), axis = 1)
    
```
